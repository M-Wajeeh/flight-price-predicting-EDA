```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df=pd.read_excel('flight_price.xlsx')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Feature Engineering 
#Seperating Date Of Journey to Date Month and Year plus converting Datatype to integer
df['Date']=df['Date_of_Journey'].str.split('/').str[0]
df['Month']=df['Date_of_Journey'].str.split('/').str[1]
df['Year']=df['Date_of_Journey'].str.split('/').str[2]
```

```python
df.head()
```

```python
df.info()
```

```python
#chaging datatype of Date Month and Year from Object to Int
df['Date']=df['Date'].astype(int)
df['Month']=df['Month'].astype(int)
df['Year']=df['Year'].astype(int)
```

```python
df.info()
```

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.columns
```

```python
#Seperating Arrival date 
df['Arrival_Time']=df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
# Splitting Arrival time into hours and minutes
df['Arrival_Hour']=df['Arrival_Time'].str.split(':').str[0]
df['Arrival_Minute']=df['Arrival_Time'].str.split(':').str[1]
```

```python
df.head(2)
```

```python
#Change Dtype from Obj to int 
df['Arrival_Hour']=df['Arrival_Hour'].astype(int)
df['Arrival_Minute']=df['Arrival_Minute'].astype(int)
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head(2)
```

```python
# similarly splitting Departure time
df['Departure_Hour']=df['Dep_Time'].str.split(':').str[0]
df['Departure_Minute']=df['Dep_Time'].str.split(':').str[1]
```

```python
df['Departure_Hour']=df['Departure_Hour'].astype(int)
df['Departure_Minute']=df['Departure_Minute'].astype(int)
```

```python
df.head(2)
```

```python
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.info()
```

```python
# Checking Categorical data instead of Numerical
df.Total_Stops.unique()
```

```python
#Assigning Labels and replace nan value with 1 (Mode of column)
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'1 stop':1 , '2 stops': 2, '3 stops':3,'4 stops':4,np.nan:1})
```

```python
df.Total_Stops.unique()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
# Dropping Route because it is showing source and destination and we already have those in seperate columns
df.drop('Route',axis=1,inplace=True)
```

```python
df['Duration_Hour']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration_Minute'] = df['Duration'].str.extract(r'(\d+)m')  # extract minutes
df['Duration_Minute'] = df['Duration_Minute'].fillna(0).astype(int)

```

```python
df.drop('Duration',axis=1,inplace=True)
```
